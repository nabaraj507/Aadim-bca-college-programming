## Instance variable and Class Variable

- Instance varibale belong to particular objects
- Class variable is shared by all object

````python
class User:

    platform = "Facebook" ## this is class variable

    def __init__(self, username): ## here username is object variable
        self.username = username

    def show(self):
        print(f"{self.username} is using {User.platform}")

user1 = User("Nabaraj")
user2 = User("Ajib")
user3 = User("Kushal")

user1.show()
user2.show()
user3.show()
````

````python
class User:
    total_user = 0

    def __init__(self, username):
        self.username = username
        User.total_user += 1
    
    def show_user(self):
        print(self.username)

    @classmethod
    def show_total(cls):
        print(cls.total_user)

user1 = User("Nabaraj")
user2 = User("Ajib")
user3 = User("Kushal")
user4 = User("Roshan")
user5 = User("Rajib")


User.show_total()
````
