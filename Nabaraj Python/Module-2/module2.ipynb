## Instance variable and Class Variable

- Instance varibale belong to particular objects
- Class variable is shared by all object

````python
class User:

    platform = "Facebook" ## this is class variable

    def __init__(self, username): ## here username is object variable
        self.username = username

    def show(self):
        print(f"{self.username} is using {User.platform}")

user1 = User("Nabaraj")
user2 = User("Ajib")
user3 = User("Kushal")

user1.show()
user2.show()
user3.show()
````

````python
class User:
    total_user = 0

    def __init__(self, username):
        self.username = username
        User.total_user += 1
    
    def show_user(self):
        print(self.username)

    @classmethod
    def show_total(cls):
        print(cls.total_user)

user1 = User("Nabaraj")
user2 = User("Ajib")
user3 = User("Kushal")
user4 = User("Roshan")
user5 = User("Rajib")


User.show_total()
````

### 1. Encapsulation:

- It is wrapping up data(variable) and member functions into a single unit and restricting direct access of Data

### Real world Example
Think about bank account

- balance(private)
- account_no(protected)
- deposit() and withdraw()


### Without Encapsulation


````python
class BankAccount:
    balance = 0

obj=BankAccount()
obj.balance = 10000
print(f"Balance is: {obj.balance}")



### balance is accessible to anyone(modify)
````

````python
class BankAccount:
    __balance = 0
    def __init__(self, acc_no, name, balance):
        self.acc_no=acc_no
        self.name = name
        self.__balance=balance    ##It is private(__, double underscore for the private)
    
    
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Rs {amount} has been deposited")



    def withdraw(self, amount):
        if(amount < 0 or self.__balance < amount):
            print("Insufficient balance.")
        else:
            self.__balance -= amount
            print(f"Rs {amount} has been withdrawn")
    

    def display(self):
        print(f"Balance is: {self.__balance}")
    


account = BankAccount("ABC123", "Nabaraj Shrestha", 10000)


account.display()
account.deposit(20000)
account.display()


account.withdraw(5000)
account.display()
````
