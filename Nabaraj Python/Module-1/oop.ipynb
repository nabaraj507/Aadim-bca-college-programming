## class

- class is a blue print to create in object.
- A class is defined as a **set of attributes and method** that created object can have.



## how to create class?
1. classes are created using keyword **class**.
2. attributes are the **set of variable** that belongs to class.
3. attributes are bydefult **public** and are used  using **dot (.)** operator>

`````python
class Dog:
    # attributes
    species ="Germen shepard"

    ## constructor
    def __init__(self,name,age):
        self.name = name 
        self.age = age
          
    def Bark(self):
          print("dog bark")

    def eat(self):
        print ("dog eat meat")
`````

 create a class name Employee with name and salary . then create a class manager,developer, which inherits Employee with department.

`````python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

class Manager(Employee):
    def __init__(self, name, salary, department):
        super().__init__(name, salary)
        self.department = department

    def display(self):
       print("Name:", self.name)
       print("Salary:", self.salary)
       print("Department:", self.department)

class Developer(Employee):
    def __init__(self, name, salary, role):
        super().__init__(name, salary)
        self.role = role 

    def display(self):
        print("Name:", self.name)
        print("salary:", self.salary)
        print("role:", self.role)

m1 = Manager("Nabaraj", 50000, "IT")
m2 = Manager("Kushal", 60000, "HR")
    
d1 = Developer("Nabaraj", 40000, "Backend")   
d2 = Developer("Kushal", 45000, "fullstack")


m1.display()   
m2.display()

d1.display()
d2.display()
`````

## Types of Inheritance 

## 1. single inheritance :
 - parent has single child 
Employee(parent class) ->  Manager (child class)

`````python
class BankAccount:
    def bank_info (self):
        print ("welcome to rastra bank")
class Account(BankAccount):
    def account_info(self):
        print("account created successfully")


# object 
acc1 = Account()
acc1.bank_info()
acc1.account_info()        
`````

## Multiple Inheritance 

- the type of inheritance where a multiple parent class has single child class.

`````python
class engine:
    def engine_type(self):
        print(" engine is atype petrol /diesel")
class whells:
        def while_cound(self):
            print("wheels : 4")
        
class car(engine,whells):
        def car_info(self):
            
            
            print("this is a car")


car1 = car()
car1.engine_type()           
car1.while_cound()
car1.car_info()           
`````

##  2 Multilevel Inheritance 

- it is the type of inheritance where feature of the parent  class and the derive class are forther inheritance into the new child class.

`````python
class Bank:
    def bank_info(self):
        print("welcome to rastra bank")
class Account(Bank):
    def account_info(self):
        print("account created ")

class Saving(Account):
    def saving_info(self):
        print("saving account with  interest")     


s = Saving()
s.bank_info()
s.account_info()
s.saving_info()                   
`````
