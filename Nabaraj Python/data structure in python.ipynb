## create a list containing student name, marks , age
operation:
- display the list 
- access last element of the list
- remove the name ram from the list 
- add sita to the list 
- add duplicate entries for age 25
- modify sita with hari 

```python
#  list containing student name, marks, age
students = [
    ["Ram", 80, 20],
    ["Shyam", 75, 22],
    ["Gita", 90, 25]
]

# display the list
print("Original list:")
print(students)

# access last element of the list
print("\nLast element:")
print(students[-1])

# remove the name Ram from the list
students = [s for s in students if s[0] != "Ram"]
print("\nAfter removing Ram:")
print(students)

# add Sita to the list
students.append(["Sita", 85, 23])
print("\nAfter adding Sita:")
print(students)

# add duplicate entries for age 25
students.append(["Duplicate1", 70, 25])
students.append(["Duplicate2", 60, 25])
print("\nAfter adding duplicate age 25 entries:")
print(students)

# modify Sita with Hari
for s in students:
    if s[0] == "Sita":
        s[0] = "Hari"

print("\nAfter modifying Sita to Hari:")
print(students)
```

```python
# Create a list of five numbers and print it
nums = [5, 2, 9, 1, 7]
print("Original list:", nums)

# Add a new element to the list
nums.append(10)
print("After adding element:", nums)

# Insert a value at a specific index
nums.insert(2, 100)
print("After inserting at index 2:", nums)

# Remove an element from the list
nums.remove(9)
print("After removing 9:", nums)

# Find the length of the list
print("Length:", len(nums))

# Access the first and last element
print("First element:", nums[0])
print("Last element:", nums[-1])

# Update an element in the list
nums[1] = 50
print("After updating index 1:", nums)

# Sort the list in ascending order
nums.sort()
print("Ascending order:", nums)

# Sort the list in descending order
nums.sort(reverse=True)
print("Descending order:", nums)

# Reverse the list
nums.reverse()
print("Reversed list:", nums)

# Find the maximum value
print("Maximum value:", max(nums))

# Find the minimum value
print("Minimum value:", min(nums))

# Find the sum of all elements
print("Sum:", sum(nums))

# Count how many times a value appears
print("Count of 10:", nums.count(10))

# Check if a value exists in the list
print("Is 7 in list?", 7 in nums)

# Copy a list into another list
copy_list = nums.copy()
print("Copied list:", copy_list)

# Merge two lists
list2 = [20, 30]
merged = nums + list2
print("Merged list:", merged)

# Remove duplicate elements from a list
dup_list = [1, 2, 2, 3, 4, 4, 5]
unique = list(set(dup_list))
print("Without duplicates:", unique)

# Find the second largest number
nums2 = [10, 20, 5, 40, 30]
nums2.sort()
print("Second largest:", nums2[-2])

# Create list of squares from 1 to 10
squares = [x**2 for x in range(1, 11)]
print("Squares:", squares)
```

## Dictionary

- it is the data structure in python which store element in **key value pair**.
- it is mostly used in **datascience and machine learning**.

'''

person= {
    "id":101,
    "name":"roshan",
    "surname":"kafle",
    "mark":60
}

'''

```python
my_dict={}
print(type(my_dict))


person= {
    "id":101,
    "name":"roshan",
    "surname":"kafle",
    "mark":60
    
}

print(person)
print(type(person))

## accessing element

print(person["id"])
print(person.get("name"))
person["surname"]="dahal"

person["age"]="10"

print(person)


## deleting element 
del person["mark"]
print(person)

val= person.pop("id")
print(val)

print(person)


## using loop 
print("keys:")
for key in person:
    print(key,end=" ")
    print()

print("display key value pair using loop")
for key, value in person.items():
    print(f"{key:}{value}")
        
```

## Task 4: 

Create a dictionary with student name, age, and course.
Example
student = {
"name": "Ram",
"age": 22,
"course": "Computer Science"
}
1. Print the dictionary.
2. Add a new key called "grade" with value "A".
3. Update age to 23.
4. Print only the course.
5. Remove the grade key.
6. Print all keys.
7. rint all values.
8. Check if the key "name" exists.
9. Print the length of the dictionary.

```python
# Create a dictionary
student = {
    "name": "Ram",
    "age": 22,
    "course": "Computer Science"
}

# 1. Print the dictionary
print("Original dictionary:", student)

# 2. Add a new key "grade"
student["grade"] = "A"
print("After adding grade:", student)

# 3. Update age to 23
student["age"] = 23
print("After updating age:", student)

# 4. Print only the course
print("Course:", student["course"])

# 5. Remove the grade key
student.pop("grade")
print("After removing grade:", student)

# 6. Print all keys
print("Keys:", student.keys())

# 7. Print all values
print("Values:", student.values())

# 8. Check if "name" exists
if "name" in student:
    print("Key 'name' exists")
else:
    print("Key 'name' does not exist")

# 9. Print the length of dictionary
print("Length of dictionary:", len(student))
```
